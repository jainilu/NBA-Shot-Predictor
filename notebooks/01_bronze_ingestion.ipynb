```python
from pyspark.sql.functions import current_timestamp, col

raw_path = "/Volumes/nba_shot_log/default/shot_log/nba_shot_logs.csv"

bronze_df = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .csv(raw_path)
)
```

```python
bronze_df = (
    bronze_df
    .withColumn("ingestion_timestamp", current_timestamp())
    .withColumn("source_file", col("_metadata.file_path"))
)
```

```python
display(bronze_df.limit(10))
```

```python
bronze_df.printSchema()
```

```python
bronze_df.write.mode("overwrite").saveAsTable("bronze_nba_shot_logs")
```
